```python
import matplotlib.pyplot as plt

def calc_median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    index_tengah = n // 2  # Pembagian bulat (floor division)

    # Jika jumlah data genap, hitung rata-rata dari dua nilai tengah
    if n % 2 == 0:
        median1 = sorted_data[index_tengah]
        median2 = sorted_data[index_tengah - 1]
        median_value = (median1 + median2) / 2
    # Jika jumlah data ganjil, ambil nilai tengah
    else:
        median_value = sorted_data[index_tengah]
    
    print(f"Median: {median_value}")
    return median_value

# Asumsikan mean_A dan mean_B sudah dihitung sebelumnya
mean_A = sum(penjualan_A) / len(penjualan_A)  # Contoh hitung mean
mean_B = sum(penjualan_B) / len(penjualan_B)  # Contoh hitung mean

# Panggil fungsi calc_median untuk kedua dataset
median_A = calc_median(penjualan_A)
median_B = calc_median(penjualan_B)

# Visualisasi Data dengan Median
plt.figure(figsize=(10, 6))

# Plot distribusi Penjualan A dengan garis Mean dan Median
plt.subplot(1, 2, 1)
plt.hist(penjualan_A, bins=10, color='blue', alpha=0.7)
plt.axvline(mean_A, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_A:.2f}')
plt.axvline(median_A, color='green', linestyle='dotted', linewidth=1, label=f'Median: {median_A:.2f}')
plt.title('Distribusi Penjualan A')
plt.xlabel('Penjualan A (pcs)')
plt.ylabel('Frekuensi')
plt.legend()

# Plot distribusi Penjualan B dengan garis Mean dan Median
plt.subplot(1, 2, 2)
plt.hist(penjualan_B, bins=10, color='green', alpha=0.7)
plt.axvline(mean_B, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_B:.2f}')
plt.axvline(median_B, color='green', linestyle='dotted', linewidth=1, label=f'Median: {median_B:.2f}')
plt.title('Distribusi Penjualan B')
plt.xlabel('Penjualan B (pcs)')
plt.ylabel('Frekuensi')
plt.legend()

# Tampilkan visualisasi
plt.tight_layout()
plt.show()
```

```python
from collections import Counter

def calc_modus(data):
    counter = Counter(data)
    mode_data = dict(counter)  # Ubah ke tipe data dictionary
    max_counter = max(list(counter.values()))  # Cari frekuensi tertinggi
    mode_value = []
    
    # Cari nilai yang memiliki frekuensi paling tinggi
    for k, v in mode_data.items():
        if v == max_counter:
            mode_value.append(k)
    
    print(f"Modus: {mode_value}")
    return mode_value

# Panggil fungsi calc_modus untuk kedua dataset
modus_A = calc_modus(penjualan_A)
modus_B = calc_modus(penjualan_B)

# Visualisasi Data dengan Mean, Median, dan Modus
plt.figure(figsize=(12, 6))

# Plot distribusi Penjualan A dengan garis Mean, Median, dan Modus
plt.subplot(1, 2, 1)  # 1 baris, 2 kolom, plot 1
plt.hist(penjualan_A, bins=10, color='blue', alpha=0.7)
plt.axvline(mean_A, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_A:.2f}')
plt.axvline(median_A, color='green', linestyle='dotted', linewidth=1, label=f'Median: {median_A:.2f}')
for mode in modus_A:
    plt.axvline(mode, color='purple', linestyle='solid', linewidth=1, label=f'Modus: {mode}')
plt.title('Distribusi Penjualan A')
plt.xlabel('Penjualan A (pcs)')
plt.ylabel('Frekuensi')
plt.legend()

# Plot distribusi Penjualan B dengan garis Mean, Median, dan Modus
plt.subplot(1, 2, 2)  # 1 baris, 2 kolom, plot 2
plt.hist(penjualan_B, bins=10, color='green', alpha=0.7)
plt.axvline(mean_B, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_B:.2f}')
plt.axvline(median_B, color='green', linestyle='dotted', linewidth=1, label=f'Median: {median_B:.2f}')
for mode in modus_B:
    plt.axvline(mode, color='purple', linestyle='solid', linewidth=1, label=f'Modus: {mode}')
plt.title('Distribusi Penjualan B')
plt.xlabel('Penjualan B (pcs)')
plt.ylabel('Frekuensi')
plt.legend()

# Tampilkan visualisasi
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt  # untuk visualisasi data

# Fungsi untuk menghitung range
def calc_range(data):
    nilai_tertinggi = max(data)
    nilai_terendah = min(data)
    data_range = nilai_tertinggi - nilai_terendah
    print(f"Range: {nilai_tertinggi} - {nilai_terendah} = {data_range}")
    return data_range

# Fungsi visualisasi untuk dua dataset
def visualize_range(penjualan_A, penjualan_B):
    # Membuat histogram untuk Penjualan A dan Penjualan B
    plt.hist(penjualan_A, bins=30, edgecolor='black', alpha=0.5, color='green', label='Penjualan A')
    plt.hist(penjualan_B, bins=30, edgecolor='black', alpha=0.5, color='orange', label='Penjualan B')

    # Menghitung range Penjualan A dan B
    calc_range(penjualan_A)
    calc_range(penjualan_B)

    # Menambahkan garis vertikal untuk nilai min dan max dari Penjualan A
    plt.axvline(x=min(penjualan_A), color='blue', linestyle='dashed', linewidth=1, label=f'Min A: {min(penjualan_A):.2f}')
    plt.axvline(x=max(penjualan_A), color='blue', linestyle='dashed', linewidth=1, label=f'Max A: {max(penjualan_A):.2f}')

    # Menambahkan garis vertikal untuk nilai min dan max dari Penjualan B
    plt.axvline(x=min(penjualan_B), color='red', linestyle='dashed', linewidth=1, label=f'Min B: {min(penjualan_B):.2f}')
    plt.axvline(x=max(penjualan_B), color='red', linestyle='dashed', linewidth=1, label=f'Max B: {max(penjualan_B):.2f}')

    # Menambahkan label, judul, dan legenda
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Range Penjualan A dan B')
    plt.legend()

    # Menampilkan plot
    plt.show()

# Contoh data penjualan
penjualan_A = dataset['Penjualan A (pcs)']
penjualan_B = dataset['Penjualan B (pcs)']

# Memanggil visualisasi
visualize_range(penjualan_A, penjualan_B)
```

```python
def calc_mean(penjualan_A):
    # Menghitung rata-rata
    return sum(penjualan_A) / len(penjualan_A)

def calc_variance(penjualan_A):
    mean = calc_mean(penjualan_A)
    
    # Kuadrat selisih antara setiap nilai data dengan rata-rata
    squared_diff = []
    for x in penjualan_A:
        squared_diff_per_x = (x - mean) ** 2
        squared_diff.append(squared_diff_per_x)
    
    # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data
    variance = sum(squared_diff) / len(penjualan_A)
    
    print(f"Variansi Penjualan A: {sum(squared_diff)} / {len(penjualan_A)} = {variance}")
    return variance

# Asumsikan dataset sudah terdefinisi sebelumnya
penjualan_B = dataset['Penjualan A (pcs)']
calc_variance(penjualan_A)
```

```python
def calc_variance_penjualan_B(penjualan_B):
    mean = calc_mean(penjualan_B)
    
    # Kuadrat selisih antara setiap nilai data dengan rata-rata
    squared_diff = []
    for x in penjualan_B:
        squared_diff_per_x = (x - mean) ** 2
        squared_diff.append(squared_diff_per_x)
    
    # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data
    variance = sum(squared_diff) / len(penjualan_B)
    
    print(f"Variansi Penjualan B: {sum(squared_diff)} / {len(penjualan_B)} = {variance}")
    return variance

# Asumsikan dataset sudah terdefinisi sebelumnya
penjualan_B = dataset['Penjualan B (pcs)']
calc_variance_penjualan_B(penjualan_B)
```

```python
import pandas as pd
from scipy.stats import kurtosis

# Membaca file Excel
dataset = pd.read_excel("Dataset.xlsx")

# Mengonversi kolom Penjualan A dan Penjualan B menjadi numerik
penjualan_A = pd.to_numeric(dataset['Penjualan A (pcs)'], errors='coerce')
penjualan_B = pd.to_numeric(dataset['Penjualan B (pcs)'], errors='coerce')

# Menghilangkan nilai NaN
penjualan_A = penjualan_A.dropna()
penjualan_B = penjualan_B.dropna()

# Menghitung Q1, Q3, dan IQR
Q1_A = penjualan_A.quantile(0.25)
Q3_A = penjualan_A.quantile(0.75)
IQR_A = Q3_A - Q1_A

Q1_B = penjualan_B.quantile(0.25)
Q3_B = penjualan_B.quantile(0.75)
IQR_B = Q3_B - Q1_B

# Menghitung batas bawah dan atas
batas_bawah_A = Q1_A - 1.5 * IQR_A
batas_atas_A = Q3_A + 1.5 * IQR_A

batas_bawah_B = Q1_B - 1.5 * IQR_B
batas_atas_B = Q3_B + 1.5 * IQR_B

# Menampilkan batas atas dan bawah untuk verifikasi
print(f"Batas bawah Penjualan A: {batas_bawah_A}, Batas atas Penjualan A: {batas_atas_A}")
print(f"Batas bawah Penjualan B: {batas_bawah_B}, Batas atas Penjualan B: {batas_atas_B}")

# Mencari outlier
outlier_A = penjualan_A[(penjualan_A < batas_bawah_A) | (penjualan_A > batas_atas_A)]
outlier_B = penjualan_B[(penjualan_B < batas_bawah_B) | (penjualan_B > batas_atas_B)]

# Menampilkan outlier dengan penanganan jika tidak ada outlier
if not outlier_A.empty:
    print("Outlier Penjualan A (pcs):")
    print(outlier_A)
else:
    print("Tidak ada outlier di Penjualan A.")

if not outlier_B.empty:
    print("Outlier Penjualan B (pcs):")
    print(outlier_B)
else:
    print("Tidak ada outlier di Penjualan B.")
```

1. Jelaskan apa fungsi dari boxplot
â€¢	Boxplot dapat mengidentifikasi rentang data: box plot menunjukkan rentang antara nilai minimum dan maksimum dalam suatu set data, memberikan pemahaman cepat tentang seberapa luas data tersebut tersebar.
â€¢	Boxplot dapat memvisualisasikan kuartil: box plot membagi data menjadi empat bagian (kuartil), sehingga memudahkan untuk melihat median (kuartil kedua) dan membandingkan kuartil pertama dan ketiga. Hal ini membantu dalam memahami seberapa terpusat atau tersebar data tersebut.
â€¢	Boxplot dapat mendeteksi outlier: dengan 'whiskers' atau garis yang membentang dari kotak, box plot menunjukkan variasi di luar kuartil atas dan bawah. Data yang berada di luar whiskers sering dianggap sebagai outlier, memberikan indikasi tentang adanya nilai yang sangat berbeda dari keseluruhan data.
â€¢	Boxplot dapat membandingkan distribusi: box plot juga berguna untuk membandingkan distribusi antara beberapa kelompok atau kategori. Hal ini memudahkan dalam melihat perbedaan median, rentang, dan variasi antara kelompok-kelompok tersebut.
