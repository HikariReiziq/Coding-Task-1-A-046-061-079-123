```python
import matplotlib.pyplot as plt  # untuk visualisasi data

# Asumsi 'dataset' sudah didefinisikan sebelumnya dan berisi kolom 'Penjualan A (pcs)' dan 'Penjualan B (pcs)'
penjualan_B, penjualan_B = dataset['Penjualan A (pcs)']
penjualan_B = dataset['Penjualan B (pcs)']

def calc_mean(penjualan_A, penjualan_B):
    # Hitung mean masing-masing variabel
    mean_A = sum(penjualan_A) / len(penjualan_A)
    mean_B = sum(penjualan_B) / len(penjualan_B)
    
    # Tampilkan hasil
    print(f"Mean Penjualan A: {mean_A}")
    print(f"Mean Penjualan B: {mean_B}")
    
    # Kembalikan hasil jika diperlukan
    return mean_A, mean_B

# Panggil fungsi calc_mean dengan dua variabel
mean_A, mean_B = calc_mean(penjualan_B, penjualan_B, penjualan_B)

# Visualisasi Data
plt.figure(figsize=(10, 6))

# Plot distribusi Penjualan A
plt.subplot(1, 2, 1)  # 1 baris, 2 kolom, plot 1
plt.hist(penjualan_B, penjualan_B, bins=10, color='blue', alpha=0.7)
plt.axvline(mean_A, color='red', linestyle='dashed', linewidth=1)
plt.title('Distribusi Penjualan A')
plt.xlabel('Penjualan A (pcs)')
plt.ylabel('Frekuensi')
plt.legend([f'Mean: {mean_A:.2f}'])

# Plot distribusi Penjualan B
plt.subplot(1, 2, 2)  # 1 baris, 2 kolom, plot 2
plt.hist(penjualan_B, bins=10, color='green', alpha=0.7)
plt.axvline(mean_B, color='red', linestyle='dashed', linewidth=1)
plt.title('Distribusi Penjualan B')
plt.xlabel('Penjualan B (pcs)')
plt.ylabel('Frekuensi')
plt.legend([f'Mean: {mean_B:.2f}'])

# Tampilkan visualisasi
plt.tight_layout()
plt.show()
```

```python
def calc_median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    index_tengah = n // 2  # Pembagian bulat (floor division)

    # Jika jumlah data genap, hitung rata-rata dari dua nilai tengah
    if n % 2 == 0:
        median1 = sorted_data[index_tengah]
        median2 = sorted_data[index_tengah - 1]
        median_value = (median1 + median2) / 2
    # Jika jumlah data ganjil, ambil nilai tengah
    else:
        median_value = sorted_data[index_tengah]
    
    print(f"Median: {median_value}")
    return median_value

# Panggil fungsi calc_median untuk kedua dataset
median_A = calc_median(penjualan_B, penjualan_B)
median_B = calc_median(penjualan_B)

# Visualisasi Data dengan Median
plt.figure(figsize=(10, 6))

# Plot distribusi Penjualan A dengan garis Mean dan Median
plt.subplot(1, 2, 1)  # 1 baris, 2 kolom, plot 1
plt.hist(penjualan_B, penjualan_B, bins=10, color='blue', alpha=0.7)
plt.axvline(mean_A, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_A:.2f}')
plt.axvline(median_A, color='green', linestyle='dotted', linewidth=1, label=f'Median: {median_A:.2f}')
plt.title('Distribusi Penjualan A')
plt.xlabel('Penjualan A (pcs)')
plt.ylabel('Frekuensi')
plt.legend()

# Plot distribusi Penjualan B dengan garis Mean dan Median
plt.subplot(1, 2, 2)  # 1 baris, 2 kolom, plot 2
plt.hist(penjualan_B, bins=10, color='green', alpha=0.7)
plt.axvline(mean_B, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_B:.2f}')
plt.axvline(median_B, color='green', linestyle='dotted', linewidth=1, label=f'Median: {median_B:.2f}')
plt.title('Distribusi Penjualan B')
plt.xlabel('Penjualan B (pcs)')
plt.ylabel('Frekuensi')
plt.legend()

# Tampilkan visualisasi
plt.tight_layout()
plt.show()
```

```python
from collections import Counter

def calc_modus(data):
    counter = Counter(data)
    mode_data = dict(counter)  # Ubah ke tipe data dictionary
    max_counter = max(list(counter.values()))  # Cari frekuensi tertinggi
    mode_value = []
    
    # Cari nilai yang memiliki frekuensi paling tinggi
    for k, v in mode_data.items():
        if v == max_counter:
            mode_value.append(k)
    
    print(f"Modus: {mode_value}")
    return mode_value

# Panggil fungsi calc_modus untuk kedua dataset
modus_A = calc_modus(penjualan_B, penjualan_B)
modus_B = calc_modus(penjualan_B)

# Visualisasi Data dengan Mean, Median, dan Modus
plt.figure(figsize=(12, 6))

# Plot distribusi Penjualan A dengan garis Mean, Median, dan Modus
plt.subplot(1, 2, 1)  # 1 baris, 2 kolom, plot 1
plt.hist(penjualan_B, penjualan_B, bins=10, color='blue', alpha=0.7)
plt.axvline(mean_A, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_A:.2f}')
plt.axvline(median_A, color='green', linestyle='dotted', linewidth=1, label=f'Median: {median_A:.2f}')
for mode in modus_A:
    plt.axvline(mode, color='purple', linestyle='solid', linewidth=1, label=f'Modus: {mode}')
plt.title('Distribusi Penjualan A')
plt.xlabel('Penjualan A (pcs)')
plt.ylabel('Frekuensi')
plt.legend()

# Plot distribusi Penjualan B dengan garis Mean, Median, dan Modus
plt.subplot(1, 2, 2)  # 1 baris, 2 kolom, plot 2
plt.hist(penjualan_B, bins=10, color='green', alpha=0.7)
plt.axvline(mean_B, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_B:.2f}')
plt.axvline(median_B, color='green', linestyle='dotted', linewidth=1, label=f'Median: {median_B:.2f}')
for mode in modus_B:
    plt.axvline(mode, color='purple', linestyle='solid', linewidth=1, label=f'Modus: {mode}')
plt.title('Distribusi Penjualan B')
plt.xlabel('Penjualan B (pcs)')
plt.ylabel('Frekuensi')
plt.legend()

# Tampilkan visualisasi
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt  # untuk visualisasi data
from collections import Counter  # untuk menghitung modus


penjualan_B = dataset['Penjualan A (pcs)']

def calc_range(penjualan_A):
    nilai_tertinggi = max(penjualan_A)
    nilai_terendah = min(penjualan_A)
    data_range = max(penjualan_A) - min(penjualan_A)
    
    print(f"Range Penjualan A: {nilai_tertinggi} - {nilai_terendah} = {data_range}")  
    return data_range


calc_range(penjualan_B)

#--------------------------------------------------------------------------------------------------------------#
penjualan_B = dataset['Penjualan B (pcs)']

def calc_range(penjualan_B):
    nilai_tertinggi = max(penjualan_B)
    nilai_terendah = min(penjualan_B)
    data_range = max(penjualan_B) - min(penjualan_B)
    
    print(f"Range Penjualan B: {nilai_tertinggi} - {nilai_terendah} = {data_range}")  
    return data_range

calc_range(penjualan_B)

#--------------------------------------------------------------------------------------------------------------#


# Fungsi visualisasi untuk dua dataset
def visualize_range(penjualan_A, penjualan_B):
    # Membuat histogram untuk Penjualan A dan Penjualan B
    plt.hist(penjualan_A, bins=30, edgecolor='black', alpha=0.5, color='green', label='Penjualan A')
    plt.hist(penjualan_B, bins=30, edgecolor='black', alpha=0.5, color='orange', label='Penjualan B')

    # Menghitung range Penjualan A dan B
    calc_range(penjualan_A)
    calc_range(penjualan_B)

    # Menambahkan garis vertikal untuk nilai min dan max dari Penjualan A
    plt.axvline(x=min(penjualan_A), color='blue', linestyle='dashed', linewidth=1, label=f'Min A: {min(penjualan_A):.2f}')
    plt.axvline(x=max(penjualan_A), color='blue', linestyle='dashed', linewidth=1, label=f'Max A: {max(penjualan_A):.2f}')

    # Menambahkan garis vertikal untuk nilai min dan max dari Penjualan B
    plt.axvline(x=min(penjualan_B), color='red', linestyle='dashed', linewidth=1, label=f'Min B: {min(penjualan_B):.2f}')
    plt.axvline(x=max(penjualan_B), color='red', linestyle='dashed', linewidth=1, label=f'Max B: {max(penjualan_B):.2f}')

    # Menambahkan label, judul, dan legenda
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Range Penjualan A dan B')
    plt.legend()

    # Menampilkan plot
    plt.show()

# Contoh data penjualan
penjualan_B = dataset['Penjualan A (pcs)']
penjualan_B = dataset['Penjualan B (pcs)']

# Memanggil visualisasi
visualize_range(penjualan_B, penjualan_B)


```

```python
def calc_mean(penjualan_A):
    # Menghitung rata-rata
    return sum(penjualan_A) / len(penjualan_A)

def calc_variance(penjualan_A):
    mean = calc_mean(penjualan_A)
    
    # Kuadrat selisih antara setiap nilai data dengan rata-rata
    squared_diff = []
    for x in penjualan_A:
        squared_diff_per_x = (x - mean) ** 2
        squared_diff.append(squared_diff_per_x)
    
    # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data
    variance = sum(squared_diff) / len(penjualan_A)
    
    print(f"Variansi Penjualan A: {sum(squared_diff)} / {len(penjualan_A)} = {variance}")
    return variance

# Asumsikan dataset sudah terdefinisi sebelumnya
penjualan_B = dataset['Penjualan A (pcs)']
calc_variance(penjualan_B)
```

```python
def calc_variance_penjualan_B(penjualan_B):
    mean = calc_mean(penjualan_B)
    
    # Kuadrat selisih antara setiap nilai data dengan rata-rata
    squared_diff = []
    for x in penjualan_B:
        squared_diff_per_x = (x - mean) ** 2
        squared_diff.append(squared_diff_per_x)
    
    # Jumlah kuadrat selisih antara setiap nilai data dengan rata-rata, kemudian dibagi dengan jumlah total data
    variance = sum(squared_diff) / len(penjualan_B)
    
    print(f"Variansi Penjualan B: {sum(squared_diff)} / {len(penjualan_B)} = {variance}")
    return variance

# Asumsikan dataset sudah terdefinisi sebelumnya
penjualan_B = dataset['Penjualan B (pcs)']
calc_variance_penjualan_B(penjualan_B)
```

```python


# Mengonversi kolom menjadi tipe numerik
penjualan_B = pd.to_numeric(dataset['Penjualan A (pcs)'], errors='coerce')
penjualan_B = pd.to_numeric(dataset['Penjualan B (pcs)'], errors='coerce')

# Menghitung jangkauan interkuartil (IQR) untuk setiap kolom
def hitung_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    return IQR

iqr_A = hitung_iqr(penjualan_B)
iqr_B = hitung_iqr(penjualan_B)

print(f'Jangkauan Interkuartil (IQR) untuk Penjualan A: {iqr_A}')
print(f'Jangkauan Interkuartil (IQR) untuk Penjualan B: {iqr_B}')

import matplotlib.pyplot as plt
import numpy as np

# Membuat Tabel Distribusi Frekuensi
def buat_tabel_frekuensi(data, bins):
    # Menghitung frekuensi untuk setiap interval
    frekuensi, batas = np.histogram(data, bins=bins)
    tabel = pd.DataFrame({
        'Interval': [f'{batas[i]:.2f} - {batas[i+1]:.2f}' for i in range(len(batas)-1)],
        'Frekuensi': frekuensi
    })
    return tabel, batas

# Menentukan jumlah kelas (bins) untuk histogram
bins = 10

# Membuat tabel distribusi frekuensi untuk Penjualan A dan B
tabel_frekuensi_A, batas_A = buat_tabel_frekuensi(penjualan_B, bins)
tabel_frekuensi_B, batas_B = buat_tabel_frekuensi(penjualan_B, bins)

print("Tabel Distribusi Frekuensi Penjualan A:")
print(tabel_frekuensi_A)
print("\nTabel Distribusi Frekuensi Penjualan B:")
print(tabel_frekuensi_B)

# Visualisasi Histogram dan Poligon Frekuensi
plt.figure(figsize=(14, 7))

# Histogram untuk Penjualan A
plt.subplot(1, 2, 1)
plt.hist(penjualan_B, bins=bins, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogram Penjualan A')
plt.xlabel('Jumlah Penjualan (pcs)')
plt.ylabel('Frekuensi')

# Poligon Frekuensi untuk Penjualan A
titik_tengah_A = [(batas_A[i] + batas_A[i+1]) / 2 for i in range(len(batas_A)-1)]
plt.plot(titik_tengah_A, tabel_frekuensi_A['Frekuensi'], marker='o', color='blue', label='Poligon Frekuensi A')
plt.legend()

# Histogram untuk Penjualan B
plt.subplot(1, 2, 2)
plt.hist(penjualan_B, bins=bins, alpha=0.7, color='green', edgecolor='black')
plt.title('Histogram Penjualan B')
plt.xlabel('Jumlah Penjualan (pcs)')
plt.ylabel('Frekuensi')

# Poligon Frekuensi untuk Penjualan B
titik_tengah_B = [(batas_B[i] + batas_B[i+1]) / 2 for i in range(len(batas_B)-1)]
plt.plot(titik_tengah_B, tabel_frekuensi_B['Frekuensi'], marker='o', color='green', label='Poligon Frekuensi B')
plt.legend()

plt.tight_layout()
plt.show()


```
